```python
import numpy as np
import matplotlib.pyplot as plt

def forward(x, w):
    """
    Forward propagation
    :param x: ndarray, input data
    :param w: scalar, weight
    :return: prediction of Y, ouput target
    """
    z = x * w
    y_pred = 1 / (1 + np.exp(-1 * z))
    # y_pred = [1 if i > 0.5 else 0 for i in y_pred]
    return y_pred

def cost(y, y_pred):
    """
    Cost function
    :param y: ndarray, input target 
    :param y_pred: ndarray, output prediction
    :return: scalar, cost
    """
    m = len(y)
    j = np.square(y - y_pred).sum() / (2 * m)
    print(j)
    return j 

# initialize samples for logistic regression
x = np.array([2, 3, 6, 7, 9, 15, 17, 4])
y = np.array([0, 0, 0, 1, 1, 1, 0, 1])

# record the historic data for plotting
w_history = []
j_history = []

# simplify the model as z = wx, y = sigmoid(z)
for w in np.arange(-5, 5, 0.1):
    y_pred = forward(x, w)
    j = cost(y, y_pred)
    w_history.append(w)
    j_history.append(j)
    
plt.figure()
plt.title("Cost Over Weight")
plt.plot(w_history, j_history, label='J over w', color='red')
plt.xticks(np.arange(-5, 5, 1))
plt.xlabel('w')
plt.ylabel('J')
plt.legend()
plt.show()
```
